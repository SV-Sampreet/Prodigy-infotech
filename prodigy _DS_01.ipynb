```python
male_population_data = df[df["Series Code"] == "SP.POP.TOTL.MA.IN"]

male_population_sorted = male_population_data.sort_values(by="2022", ascending=False)

male_top_ten_countries = male_population_sorted.head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['Country Name']] )
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=male_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Highest Male Population (2022)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

```python
df['2022_female'] = df['2022'] * 0.495

female_top_ten_countries = df[['Country Name', '2022_female']].dropna().sort_values(
    by='2022_female', ascending=False).head(10)

female_top_ten_countries.rename(columns={'2022_female': '2022'}, inplace=True)
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(2, 2, 2)
sns.barplot(x="2022", y="Country Name", data=female_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Highest Female Population (2022)", size=10)
plt.xlabel("Female Population", size=10)
plt.ylabel("Country", size=10)
plt.tight_layout()
plt.show()
```

```python
male_population_data = df[['Country Name', '2022']].copy()
male_population_data['2022'] = male_population_data['2022'] * 0.505

female_population_data = df[['Country Name', '2022']].copy()
female_population_data['2022'] = female_population_data['2022'] * 0.495

male_population_data.rename(columns={'2022': 'Male Population'}, inplace=True)
female_population_data.rename(columns={'2022': 'Female Population'}, inplace=True)
```

```python
merged_data = pd.merge(male_population_data, female_population_data, on='Country Name')
merged_data['Total Population'] = merged_data['Male Population'] + merged_data['Female Population']
```

```python
sorted_data = merged_data.sort_values(by='Total Population', ascending=False)
top_10_countries = sorted_data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
sns.barplot(x='Total Population', y='Country Name', data=top_10_countries, palette='magma')
plt.title("Top 10 Countries by Estimated Total Population (2022)")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
